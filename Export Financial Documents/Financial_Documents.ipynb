```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from vnstock import *
dictE_V = {
    'quarter': 'Quý',
    'year': 'Năm',
    'revenue': 'Doanh thu',
    'costOfGoodSold': 'Giá vốn hàng bán',
    'grossProfit': 'Lợi nhuận gộp',
    'operationExpense': 'Chi phí hoạt động',
    'operationProfit': 'Lợi nhuận hoạt động kinh doanh',
    'interestExpense': 'Chi phí lãi vay',
    'preTaxProfit': 'Lợi nhuận trước thuế',
    'postTaxProfit': 'Lợi nhuận sau thuế',
    'shareHolderIncome': 'Lợi nhuận sau thuế và sau CĐ thiểu số',
    'ebitda': 'EBITDA',
    'shortAsset': 'Tài sản ngắn hạn',
    'cash': 'Tiền và tương đương tiền',
    'shortInvest': 'Đầu tư ngắn hạn',
    'shortReceivable': 'Phải thu khách hàng',
    'inventory': 'Hàng tồn kho',
    'longAsset': 'Tài sản dài hạn',
    'fixedAsset': 'Tài sản cố định',
    'asset': 'Tổng tài sản',
    'debt': 'Tổng nợ phải trả',
    'shortDebt': 'Vay ngắn hạn',
    'longDebt': 'Vay dài hạn',
    'equity': 'Vốn chủ sở hữu',
    'capital': 'Vốn điều lệ',
    'investCost': 'Chi phí đầu tư',
    'fromInvest': 'Từ HĐ đầu tư',
    'fromFinancial': 'Từ HĐ tài chính',
    'fromSale': 'Từ HĐ kinh doanh',
    'centralBankDeposit': 'Tiền gửi tại NHNN',
    'otherBankDeposit': 'Tiền gửi tại các TC tín dụng khác',
    'otherBankLoan': 'Cho các TCTD khác vay',
    'stockInvest': 'Chứng khoán đầu tư',
    'customerLoan': 'Cho vay khách hàng',
    'badLoan': 'Nợ xấu',
    'provision': 'Dự phòng cho vay KH',
    'netCustomerLoan': 'Cho vay KH ròng',
    'otherAsset': 'Tài sản khác',
    'otherBankCredit': 'Tiền gửi của các TCTD',
    'oweOtherBank': 'Vay các TCTD',
    'oweCentralBank': 'Vay NHNN',
    'valuablePaper': 'Phát hành giấy tờ có giá',
    'payableInterest': 'Lãi và phí phải trả',
    'receivableInterest': 'Lãi và phí phải thu',
    'deposit': 'Tiền gửi của khách hàng',
    'otherDebt': 'Nợ phải trả khác',
    'fund': 'Quỹ của TCTD',
    'unDistributedIncome': 'Lợi nhuận chưa phân phối',
    'minorShareHolderProfit': 'Lợi ích không kiểm soát',
    'investProfit': 'Lãi thuần từ hoạt động đầu tư',
    'serviceProfit': 'Lãi thuần từ hoạt động dịch vụ',
    'otherProfit': 'Lãi thuần từ hoạt động khác',
    'provisionExpense': 'Chi phí dự phòng',
    'operationIncome': 'Lợi nhuận trước dự phòng'
}

draw_list = ['Doanh thu', 'Giá vốn hàng bán', 'Lợi nhuận gộp', 'Lợi nhuận hoạt động kinh doanh', 'Lợi nhuận trước thuế', 'Lợi nhuận sau thuế',
             'Hàng tồn kho', 'Tài sản dài hạn', 'Tài sản cố định', 'Tổng tài sản', 'Tổng nợ phải trả', 'Vốn chủ sở hữu', 'Từ HĐ kinh doanh']

list_of_keys = list(dictE_V.keys())
def convert_to_year(s: str):
    parts = s.split("-") + s.split("Q")
    year = parts[0]
    return year

def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center', fontweight='bold')
```

```python
symbol = input("Stock: ")
start_year = int(input("Start year: "))
end_year = int(input("End year: "))
year_list = list(range(start_year, end_year + 1))
year_list = [str(year) for year in year_list]
```

```python
# Year
bls = financial_flow(symbol, report_type='balancesheet', report_range='yearly')
bls = bls.transpose()
bls = bls.iloc[:, ::-1]
bls = bls.rename(columns=lambda x: convert_to_year(x))
year_list = bls.columns.intersection(year_list)
bls = bls[year_list]

ins = financial_flow(symbol, report_type='incomestatement', report_range='yearly')
ins = ins.transpose()
ins = ins.iloc[:, ::-1]
ins = ins.rename(columns=lambda x: convert_to_year(x))
year_list = ins.columns.intersection(year_list)
ins = ins[year_list]

cfl = financial_flow(symbol, report_type='cashflow', report_range='yearly')
cfl = cfl.transpose()
cfl = cfl.iloc[:, ::-1]
cfl = cfl.rename(columns=lambda x: convert_to_year(x))
year_list = cfl.columns.intersection(year_list)
cfl = cfl[year_list]

df = pd.concat([bls, ins, cfl], axis=0, ignore_index=False)
df.dropna(inplace=True)
df = df.reset_index()
df = df[df['index'].isin(list_of_keys)]
df.to_excel("Output/(Eng) " + symbol + "_" + str(year_list[0]) + "_" + str(year_list[-1]) + "(Y)" + ".xlsx")
df['index'] = df['index'].map(dictE_V)
df.to_excel("Output/(Vie) " + symbol + "_" + str(year_list[0]) + "_" + str(year_list[-1]) + "(Y)" + ".xlsx")
df = df.set_index('index')
df = df.transpose()
draw_list = df.columns.intersection(draw_list)
sns.set_theme(style="whitegrid")
for col in draw_list:
    plt.figure(figsize=(24,6))
    plt.title(col, fontweight='bold', fontsize=18)
    plt.bar(df.index, df[col], color=(0.1, 0.1, 0.1, 0.1),  edgecolor='blue')
    addlabels(df.index, df[col])
    plt.savefig('Output/Img/Year/' + col, dpi=400)
    plt.close()


# Quarter
quarter_list = []
for year in year_list:
    for q in range(1,5):
        quarter_list.append(str(year + '-Q' + str(q)))
bls = financial_flow(symbol, report_type='balancesheet', report_range='quarterly')
bls = bls.transpose()
bls = bls.iloc[:, ::-1]
quarter_list = bls.columns.intersection(quarter_list)
bls = bls[quarter_list]

ins = financial_flow(symbol, report_type='incomestatement', report_range='quarterly')
ins = ins.transpose()
ins = ins.iloc[:, ::-1]
quarter_list = ins.columns.intersection(quarter_list)
ins = ins[quarter_list]

cfl = financial_flow(symbol, report_type='cashflow', report_range='quarterly')
cfl = cfl.transpose()
cfl = cfl.iloc[:, ::-1]
quarter_list = cfl.columns.intersection(quarter_list)
cfl = cfl[quarter_list]

df = pd.concat([bls, ins, cfl], axis=0, ignore_index=False)
df.dropna(inplace=True)
df = df.reset_index()
df = df[df['index'].isin(list_of_keys)]
df.to_excel("Output/(Eng) " + symbol + "_" + str(year_list[0]) + "_" + str(year_list[-1]) + "(Q)" + ".xlsx")
df['index'] = df['index'].map(dictE_V)
df.to_excel("Output/(Vie) " + symbol + "_" + str(year_list[0]) + "_" + str(year_list[-1]) + "(Q)" + ".xlsx")
df = df.set_index('index')
df = df.transpose()
draw_list = df.columns.intersection(draw_list)
for col in draw_list:
    plt.figure(figsize=(24,6))
    plt.title(col, fontweight='bold', fontsize=18)
    plt.bar(df.index, df[col], color=(0.1, 0.6, 0.1, 0.1),  edgecolor='green')
    addlabels(df.index, df[col])
    plt.savefig('Output/Img/Quarter/' + col, dpi=400)
    plt.close()
```
